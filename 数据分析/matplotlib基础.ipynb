```python
import matplotlib.pyplot as plt
import random

#图表里显示中文
plt.rcParams['font.sans-serif']=['SimHei'] 
#图表里负号显示正常
plt.rcParams['axes.unicode_minus'] = False

#设置画布大小
plt.figure(figsize=(10,5))
#设置标题
plt.title("温差变化折线图",fontsize=20)
x1 = list(range(1,16))
y1 = [random.randint(5, 30) for m in range(15)] 
x2 = list(range(1,16))
y2 = [random.randint(-5, 10) for m in range(15)] 
#x和y的数据数量要一致,label可以直接设置图例名称
plt.plot(x1,y1,label="武汉")
plt.plot(x2,y2,label="北京")
#设置图例 ，loc设置图例位置
plt.legend(["武汉","北京"],loc="upper right")
#设置x，y轴刻度  
plt.xticks(x1[::1])
plt.yticks(range(-5,31))
#设置x，y轴标签
plt.xlabel("日期")
plt.ylabel("温度（℃）")
#设置网格 ,linestyle设置线型，linewidth设置线宽，color设置线颜色
plt.grid(True, linestyle='--', linewidth='0.6', color="#000000") 
plt.savefig("d:/github/python-noob/数据分析/温差变化折线图.png")


plt.show()
```

```python
from matplotlib import axis
import matplotlib.pyplot as plt

#创建子图 
plt.subplots(nrows = 1,ncols =2,figsize = (20,6.5),dpi = 100) 

contries = ["中国","美国","俄罗斯","英国","法国"]
vlaues = [960,937,1707,55,24.4]
axis1 = plt.subplot(1,2,1)
axis2 = plt.subplot(1,2,2)
axis1.set_title("五常各国领土柱状图")
axis2.set_title("五常各国领土饼状图")

#设置柱状图
axis1.bar(contries,vlaues,color = ["#FF0000","#00FF00","#0000FF","#FFA500","#00FFFF"])
axis1.set_ylabel("国土面积（万平方公里）")
_ =axis1.set_yticks(range(0,1800,100))

#设置饼状图
axis2.pie(vlaues,labels = contries,colors = ["#FF0000","#00FF00","#0000FF","#FFA500","#00FFFF"],autopct = "%1.1f%%",shadow = False,startangle = 90)
axis2.set_xlabel("国土面积（万平方公里）",fontsize = 12)
axis2.legend()

plt.savefig('D:\github\python-noob\数据分析\五常各国领土.png')
plt.show()
```
