```python
#数据选择  
df = pd.read_csv(r'D:\github\python-noob\数据分析\shenzhen(1).csv',encoding='utf-8')

#选择列
df['总价']  #获取单列
df[['hourseSize','fitment','总价']] #获取多列


#选择行     iloc()  loc()
#iloc()------->通过行号获取数据,不包含结束的数据   df.iloc[start:end:step]
df1 = df.iloc[0:10:2]
df1

#loc()------->通过索引获取数据,包含结束的数据   df.loc[start:end:step]  当没有索引改变时,会报错
df2 = df.loc[0:10:2]
df2
```

```python
#数据筛选 df[filter]
df = pd.read_csv(r'D:\github\python-noob\数据分析\shenzhen(1).csv',encoding='utf-8')

#展示所有数据
pd.set_option('display.max_rows',None)

# #1.获取unit_price大于60000
df[df['unit_price'] > 60000]    
#2.获取community为金城华庭或洪湖东岸的
df[df['community'].isin(['金城华庭' ,'洪湖东岸'])]
#3.获取unit_price大于60000且fitment为精装的
df[(df['unit_price'] > 60000) & (df['fitment'] == '精装')]
#4.获取总价在500到600的
df[df['总价'].between (500,600)]
#5.获取community为金城华庭或洪湖东岸的，总价为500到800的
df[df['community'].isin(['金城华庭' ,'洪湖东岸']) &df['总价'].between (500,800)]
```
