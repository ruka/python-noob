```python
#数据分组
import pandas as pd
df = pd.read_csv('data02.csv')
#根据产品类别进行分组并统计产品名称数量
df.groupby('产品类别')['产品名称'].count()
#根据产品类别进行分组并统计销售数量之和
df.groupby('产品类别')['销售数量'].sum()
#根据产品类别进行分组并统计销售金额之和
df['销售金额'] = df['销售数量']*df['单价']
df.groupby('产品类别')['销售金额'].sum()
#根据产品类别进行分组并统计销售商品平均值
df.groupby('产品类别')['单价'].mean()
#根据产品类别进行分组并统计销售商品最大值
df.groupby('产品类别')['单价'].max()
#根据产品类别进行分组并统计销售商品最小值
df.groupby('产品类别')['单价'].min()
#根据产品类别进行分组并统计销售商品最小值,最大值,平均值
df.groupby('产品类别')['单价'].agg(['min','max','mean']).round(2)
df.groupby('产品类别').agg({'单价':['min','max','mean']}).round(2)
#根据产品类别进行分组并统计销售数量之和,销售金额之和，平均单价
df.groupby('产品类别').agg({'销售数量':'sum', '销售金额':'sum', '单价':'mean'}).round(2)
   
```
