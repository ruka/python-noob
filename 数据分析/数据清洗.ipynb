```python
import pandas as pd
pd.set_option('display.max_rows',20)
df = pd.read_csv(r'D:\github\python-noob\数据分析\data.csv',encoding='utf-8',index_col=0)
#缺失值处理
df.isnull()   #判断是否有缺失值
#删除缺失值
df.dropna(inplace=True)  #删除有缺失值的行
df.dropna(axis=1)  #删除有缺失值的列
# 填充缺失值 
df.fillna(99999) #填充缺失值
df.ffill()        #用上一行数据填充缺失值
df.bfill()          #用下一行数据填充缺失值
```

```python
#查找重复值
import pandas as pd
pd.set_option('display.max_rows',20)
df = pd.read_csv(r'D:\github\python-noob\数据分析\data.csv',encoding='utf-8')
#查看是否有重复的
df.duplicated()    #查找所有列是否重复
df.duplicated(subset=['ID'])   #查找ID列（索引）是否重复
# #删除重复值
df.drop_duplicates(subset=['ID'],keep='first') #保留第一次出现的
df.drop_duplicates(subset=['ID'],keep='last') #保留最后一次出现的
```

```python
#处理异常值
import pandas as pd
pd.set_option('display.max_rows',20)
df = pd.read_csv(r'D:\github\python-noob\数据分析\data.csv',encoding='utf-8')
df[(df['年龄'] < 0) | (df['工资'] < 0)] #查看异常值
df[(df['年龄'] < 0) | (df['工资'] < 0)].index #查看异常值的索引
#删除异常值
# df.drop(df[(df['年龄'] < 0) | (df['工资'] < 0)].index)
#修复异常值
# df['工资']=df['工资'].abs()
# df['年龄']=df['年龄'].abs()
df[['工资','年龄']] = df[['工资','年龄']].abs() #将两列的数据取绝对值
df
```
